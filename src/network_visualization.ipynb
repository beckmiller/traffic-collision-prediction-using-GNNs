```python
import pandas as pd

accident_df = pd.read_csv('/Users/beck/Documents/GitHub/ML4RoadSafety/CA/Accidents_Nearest_Street_CA_Monthly.csv')
node_df = pd.read_csv('/Users/beck/Documents/GitHub/ML4RoadSafety/CA/Nodes/node_features_2013_1.csv')
```

```python
# Merge accident data with node coordinates for node_1
accident_node1 = pd.merge(accident_df, node_df, left_on='node_1', right_on='node_id')
accident_node1 = accident_node1.rename(columns={'lat': 'lat1', 'long': 'long1'})

# Merge accident data with node coordinates for node_2
accident_node2 = pd.merge(accident_node1, node_df, left_on='node_2', right_on='node_id')
accident_node2 = accident_node2.rename(columns={'lat': 'lat2', 'long': 'long2'})

view_data = pd.DataFrame({
    'lat1': accident_node2['lat1'],
    'long1': accident_node2['long1'],
    'lat2': accident_node2['lat2'],
    'long2': accident_node2['long2'],
    'count': accident_node2['acc_count']
})
```

```python
import pydeck as pdk

"""
HexagonLayer
==============

Personal injury road accidents in GB from 1979.

The layer aggregates data within the boundary of each hexagon cell.

This example is adapted from the deck.gl documentation.
"""

HEXAGON_LAYER_DATA = (
    "https://raw.githubusercontent.com/visgl/deck.gl-data/master/examples/3d-heatmap/heatmap-data.csv"  # noqa
)

# Define a layer to display on a map
layer = pdk.Layer(
    "HexagonLayer",
    HEXAGON_LAYER_DATA,
    get_position=["lng", "lat"],
    auto_highlight=True,
    elevation_scale=50,
    pickable=True,
    elevation_range=[0, 3000],
    extruded=True,
    coverage=1,
)

# Set the viewport location
view_state = pdk.ViewState(
    longitude=-1.415,
    latitude=52.2323,
    zoom=6,
    min_zoom=5,
    max_zoom=15,
    pitch=40.5,
    bearing=-27.36,
)

# Render
r = pdk.Deck(layers=[layer], initial_view_state=view_state)
r.to_html("hexagon_layer.html")
```
