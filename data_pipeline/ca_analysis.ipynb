```python
# visualize road network
def concat_files(path, final_file_name):
    '''
    Combines all files in a directory and saves it in a single file
    Parameters:
        path (str): directory where all independent files are saved
        final_file_name (str): path of the final file
    '''
    count = 0
    for file_name in os.listdir(path):
        if file_name.endswith('.csv'):
            try:
                df = pd.concat([df, pd.read_csv(path + file_name, low_memory=False)])
            except:
                df = pd.read_csv(path + file_name, low_memory=False)
            count += 1
    df = df.drop_duplicates().reset_index(drop=True)
    df.to_csv(final_file_name, index=False)

```

```python
import torch

# load the adjacency matrix from a .pt file
adj_matrix = torch.load('CA-2/adj_matrix.pt')
```

```python

# assume adj_matrix is your sparse matrix
indices = adj_matrix._indices()
values = adj_matrix._values()

# number of edges
num_edges = values.shape[0]
print(f"Number of edges: {num_edges}")

# basic stats
print(f"Min value: {values.min().item()}, Max value: {values.max().item()}")

sample_indices = indices[:, :10]  # first 10 edges
sample_values = values[:10]
print("Sample edges (row, col):")
print(sample_indices)
print("Sample values:")
print(sample_values)
row_indices = indices[0, :]
col_indices = indices[1, :]

# degrees per node
# use torch.bincount for fast counting of occurrences
out_degrees = torch.bincount(row_indices, minlength=adj_matrix.size(0))
in_degrees = torch.bincount(col_indices, minlength=adj_matrix.size(1))

print(f"Average out-degree: {out_degrees.float().mean().item()}")
print(f"Average in-degree: {in_degrees.float().mean().item()}")
```

```python
import pydeck

"""
HexagonLayer
==============

Personal injury road accidents in GB from 1979.

The layer aggregates data within the boundary of each hexagon cell.

This example is adapted from the deck.gl documentation.
"""

import pydeck as pdk

HEXAGON_LAYER_DATA = (
    "https://raw.githubusercontent.com/visgl/deck.gl-data/master/examples/3d-heatmap/heatmap-data.csv"  # noqa
)

# Define a layer to display on a map
layer = pdk.Layer(
    "HexagonLayer",
    HEXAGON_LAYER_DATA,
    get_position=["lng", "lat"],
    auto_highlight=True,
    elevation_scale=50,
    pickable=True,
    elevation_range=[0, 3000],
    extruded=True,
    coverage=1,
)

# Set the viewport location
view_state = pdk.ViewState(
    longitude=-1.415,
    latitude=52.2323,
    zoom=6,
    min_zoom=5,
    max_zoom=15,
    pitch=40.5,
    bearing=-27.36,
)

# Render
r = pdk.Deck(layers=[layer], initial_view_state=view_state)
r.to_html("hexagon_layer.html")


```
